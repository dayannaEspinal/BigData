```python
### Ejercicio 17: Análisis de ventas con pandas
- Cree un DataFrame de ventas con al menos 15 registros.
- Incluya fecha, vendedor, producto, categoría, cantidad y precio unitario.
- Calcule una nueva columna TotalVenta.
- Calcule el total vendido por vendedor.
- Identifique el producto más vendido.
- Identifique la categoría con mayor ingreso.
- Ordene las ventas de mayor a menor.
```

```python
import pandas as pd
data = {
    "Fecha": pd.date_range("2024-01-01", periods=15, freq="D"),
    "Vendedor": ["Dayanna", "Ivette", "Lucas", "Dayanna", "Ivette", "Lucas", "Dayanna", "Ivette", "Lucas", "Dayanna", "Ivette", "Lucas", "Dayanna", "Ivette", "Lucas"],
    "Producto": ["Laptop", "Mouse", "Teclado", "Laptop", "Monitor", "Mouse", "Celular", "Laptop", "Teclado", "Monitor", "Celular", "Laptop", "Mouse", "Teclado", "Monitor"],
    "Categoría": ["Electrónica", "Accesorios", "Accesorios", "Electrónica", "Electrónica", "Accesorios", "Electrónica", "Electrónica", "Accesorios", "Electrónica", "Electrónica", "Electrónica", "Accesorios", "Accesorios", "Electrónica"],
    "Cantidad": [2, 5, 3, 1, 2, 4, 3, 2, 6, 1, 2, 1, 7, 3, 2],
    "PrecioUnitario": [800, 20, 50, 850, 300, 25, 600, 820, 55, 310, 620, 830, 22, 48, 305]
}
ventas = pd.DataFrame(data)
ventas["TotalVenta"] = ventas["Cantidad"] * ventas["PrecioUnitario"]
total_vendedor = ventas.groupby("Vendedor")["TotalVenta"].sum()
producto_mas_vendido = ventas.groupby("Producto")["Cantidad"].sum().idxmax()
categoria_mayor_ingreso = ventas.groupby("Categoría")["TotalVenta"].sum().idxmax()
ventas_ordenadas = ventas.sort_values(by="TotalVenta", ascending=False)
print("=== DataFrame de Ventas ===")
print(ventas)
print("\nTotal vendido por vendedor")
print(total_vendedor)
print("\nProducto más vendido")
print(producto_mas_vendido)
print("\nCategoría con mayor ingreso")
print(categoria_mayor_ingreso)
print("\nVentas ordenadas de mayor a menor")
print(ventas_ordenadas)
```

```python
##Explicación:
- Se crea un DataFrame con 15 registros de ventas.
- Se calcula TotalVenta multiplicando Cantidad * PrecioUnitario.
- Se agrupan las ventas por Vendedor para obtener el total vendido.
- Se identifica el producto más vendido usando idxmax() sobre la suma de cantidades.
- Se identifica la categoría con mayor ingreso agrupando por Categoría.
- Finalmente, se ordenan las ventas de mayor a menor según TotalVenta.
```

```python
# Ejercicio 18: Limpieza de datos con pandas
- Cree un DataFrame con datos incompletos de clientes.
- Incluya nombre, edad, correo, ciudad y teléfono. 
- Identifique valores nulos. 
- Cuente los valores nulos por columna. 
- Reemplace ciudades vacías con "No especificada". 
- Elimine registros que no tengan correo. 
- Muestre el DataFrame limpio. 
```

```python
import pandas as pd
data = {
    "Nombre": [
        "Kevin Mendoza",
        "Laura Castillo",
        "Fernando Aguilar",
        "Camila Reyes",
        "Miguel Santos",
        "Paola Rivera"
    ],

    "Edad": [22, 31, None, 27, 40, None],

    "Correo": [
        "kevin@gmail.com",
        None,
        "fernando@gmail.com",
        "camila@gmail.com",
        None,
        "paola@gmail.com"
    ],

    "Ciudad": [
        "Choluteca",
        "",
        "Tegucigalpa",
        None,
        "San Pedro Sula",
        ""
    ],

    "Telefono": [
        "9988-1122",
        "8877-3344",
        None,
        "7766-5566",
        "6655-7788",
        None
    ]
}

df = pd.DataFrame(data)

print("=" * 60)
print("DATAFRAME ORIGINAL")
print("=" * 60)
print(df)
print("\n" + "=" * 60)
print("VALORES NULOS")
print("=" * 60)
print(df.isnull())
print("\n" + "=" * 60)
print("CANTIDAD DE VALORES NULOS POR COLUMNA")
print("=" * 60)
print(df.isnull().sum())
df["Ciudad"] = df["Ciudad"].replace("", "No especificada")
df["Ciudad"] = df["Ciudad"].fillna("No especificada")
df_limpio = df.dropna(subset=["Correo"])
print("\n" + "=" * 60)
print("DATAFRAME LIMPIO")
print("=" * 60)
print(df_limpio)
```

```python
###Ejercicio 19: Análisis de préstamos
- Cree un DataFrame con solicitudes de préstamo.
- Incluya cliente, monto solicitado, plazo, ingreso mensual y estado de solicitud.
- Calcule el monto promedio solicitado.
- Cuente solicitudes por estado.
- Muestre únicamente préstamos aprobados.
- Muestre clientes con monto solicitado mayor a 50,000.
- Agrupe por estado y calcule el promedio del monto solicitado.
```

```python
import pandas as pd


datos = {
    "Cliente": ["Juan", "María", "Carlos", "Ana", "Pedro", "Lucía"],
    "Monto solicitado": [45000, 60000, 30000, 80000, 55000, 40000],
    "Plazo (meses)": [12, 24, 18, 36, 24, 12],
    "Ingreso mensual": [15000, 25000, 12000, 30000, 22000, 18000],
    "Estado": ["Aprobado", "Rechazado", "Aprobado", "Aprobado", "Pendiente", "Rechazado"]
}

prestamos = pd.DataFrame(datos)

print("Lista de solicitudes de préstamo")
print(prestamos)


promedio = prestamos["Monto solicitado"].mean()

print("\nPromedio del monto solicitado:")
print(promedio)


print("\nCantidad de solicitudes por estado:")
print(prestamos["Estado"].value_counts())


print("\nPréstamos aprobados:")
print(prestamos[prestamos["Estado"] == "Aprobado"])


print("\nClientes con préstamos mayores a 50,000:")
print(prestamos[prestamos["Monto solicitado"] > 50000])

print("\nPromedio del monto solicitado según el estado:")
print(prestamos.groupby("Estado")["Monto solicitado"].mean())
```

```python
###Ejercicio 20: Mini análisis de datos tipo empresa
- Cree un DataFrame con información de empleados.
- Incluya nombre, departamento, cargo, salario, años de experiencia y ciudad.
- Muestre los primeros registros.
- Muestre estadísticas generales de salarios.
- Calcule el salario promedio por departamento.
- Encuentre el empleado con salario más alto.
- Muestre empleados con más de 5 años de experiencia.
- Ordene empleados por salario.
- Cree una columna NivelExperiencia: Junior, Intermedio o Senior.
```

```python
import pandas as pd


data = {
    "Nombre": [
        "Ana García", "Luis Martínez", "María López", "Carlos Pérez",
        "Sofía Hernández", "Jorge Ramírez", "Daniela Torres", "Andrés Flores",
        "Valeria Morales", "Ricardo Gómez"
    ],
    "Departamento": [
        "Ventas", "TI", "RRHH", "TI",
        "Ventas", "Finanzas", "RRHH", "Finanzas",
        "TI", "Ventas"
    ],
    "Cargo": [
        "Ejecutivo de Ventas", "Desarrollador", "Analista RRHH", "Arquitecto de Software",
        "Gerente de Ventas", "Contador", "Jefe de RRHH", "Analista Financiero",
        "Desarrollador Senior", "Asesor Comercial"
    ],
    "Salario": [
        35000, 52000, 40000, 75000,
        68000, 47000, 61000, 43000,
        70000, 32000
    ],
    "AniosExperiencia": [3, 6, 4, 10, 8, 5, 9, 2, 11, 1],
    "Ciudad": [
        "Bogotá", "Medellín", "Bogotá", "Cali",
        "Medellín", "Bogotá", "Cali", "Barranquilla",
        "Medellín", "Bogotá"
    ]
}

df = pd.DataFrame(data)

print("=" * 60)
print("PRIMEROS REGISTROS")
print("=" * 60)
print(df.head())
print("\n" + "=" * 60)
print("ESTADÍSTICAS GENERALES DE SALARIOS")
print("=" * 60)
print(df["Salario"].describe())
print("\n" + "=" * 60)
print("SALARIO PROMEDIO POR DEPARTAMENTO")
print("=" * 60)
print(df.groupby("Departamento")["Salario"].mean())
print("\n" + "=" * 60)
print("EMPLEADO CON SALARIO MÁS ALTO")
print("=" * 60)
print(df.loc[df["Salario"].idxmax()])
print("\n" + "=" * 60)
print("EMPLEADOS CON MÁS DE 5 AÑOS DE EXPERIENCIA")
print("=" * 60)
print(df[df["AniosExperiencia"] > 5])
print("\n" + "=" * 60)
print("EMPLEADOS ORDENADOS POR SALARIO (MAYOR A MENOR)")
print("=" * 60)
print(df.sort_values("Salario", ascending=False))
def nivel_experiencia(anios):
    if anios <= 3:
        return "Junior"
    elif anios <= 6:
        return "Intermedio"
    else:
        return "Senior"

df["NivelExperiencia"] = df["AniosExperiencia"].apply(nivel_experiencia)
print("\n" + "=" * 60)
print("DATAFRAME CON COLUMNA NIVELEXPERIENCIA")
print("=" * 60)
print(df[["Nombre", "AniosExperiencia", "NivelExperiencia"]])
```
