```python
#Ejercicio 8: Lista de números positivos y negativos
- Defina o solicite 10 números.
- Cuente cuántos son positivos, negativos y ceros.
- Mostrar un resumen final.
```

```python
numeros = [0, 0, 50, -45, 64, 75, -32, -85, 49, 41]
positivos = 0
negativos = 0
ceros = 0
for num in numeros:
    if num > 0:
        positivos += 1
    elif num < 0:
        negativos += 1
    else:
        ceros += 1
print("=== RESUMEN DE CLASIFICACIÓN ===")
print(f"Lista de números evaluados: {numeros}")
print("-" * 33)
print(f"Cantidad de números Positivos: {positivos}")
print(f"Cantidad de números Negativos: {negativos}")
print(f"Cantidad de números Ceros:     {ceros}")
print("================================")
```

```python
=== RESUMEN DE CLASIFICACIÓN ===
Lista de números evaluados: [0, 0, 50, -45, 64, 75, -32, -85, 49, 41]
---------------------------------
Cantidad de números Positivos: 5
Cantidad de números Negativos: 3
Cantidad de números Ceros:     2
================================
```

```python
#Ejercicio 11 Función para calcular descuento
- Cree una función llamada calcular_descuento.
- La función debe recibir precio original y porcentaje de descuento.
- Debe retornar el precio final.
- Explique qué datos entran a la función y qué dato retorna.
```

```python
def calcular_descuento(precio_original, porcentaje_descuento):
    """
    Parámetros que entran:
      precio_original      → float/int  (ej: 500.00)
      porcentaje_descuento → float/int  (ej: 20  → 20%)
    Dato que retorna:
      precio_final         → float      (precio con descuento aplicado)
    """
    monto_descuento = precio_original * (porcentaje_descuento / 100)
    precio_final    = precio_original - monto_descuento
    return precio_final
precio     = 500.00
descuento  = 20
final      = calcular_descuento(precio, descuento)
print(f"Precio original  : L. {precio:.2f}")
print(f"Descuento        : {descuento}%")
print(f"Precio final     : L. {final:.2f}")
```

```python
Precio original  : L. 500.00
Descuento        : 20%
Precio final     : L. 400.00
```

```python
#Ejercicio 12 Validación de contraseña
- Valide una contraseña definida por el usuario o por el código.
- Debe tener al menos 8 caracteres.
- Debe contener al menos un número.
- Debe contener al menos una letra mayúscula.
- Debe contener al menos una letra minúscula.
- Indique si la contraseña es válida o no.
```

```python
def validar_contrasena(contrasena):
    tiene_longitud  = len(contrasena) >= 8
    tiene_numero    = any(c.isdigit()  for c in contrasena)
    tiene_mayuscula = any(c.isupper()  for c in contrasena)
    tiene_minuscula = any(c.islower()  for c in contrasena)
    print(f"  Mínimo 8 caracteres   : {'✔' if tiene_longitud  else '✘'}")
    print(f"  Contiene número       : {'✔' if tiene_numero    else '✘'}")
    print(f"  Contiene mayúscula    : {'✔' if tiene_mayuscula else '✘'}")
    print(f"  Contiene minúscula    : {'✔' if tiene_minuscula else '✘'}")

    if tiene_longitud and tiene_numero and tiene_mayuscula and tiene_minuscula:
        return "✔ Contraseña VÁLIDA"
    else:
        return "✘ Contraseña INVÁLIDA"

# Pruebas
for pwd in ["Segura1!", "débil", "SINMINUSCULA8"]:
    print(f'\nContraseña: "{pwd}"')
    print(validar_contrasena(pwd))
```

```python
# Ejercicio 13: Sistema simple de inventario
- Cree una lista de productos.
- Cada producto debe tener código, nombre, precio y cantidad disponible.   
- Muestre todos los productos.
- Busque un producto por código.
- Calcule el valor total del inventario.
```

```python
inventario = [
    {"codigo": "P001", "nombre": "Laptop", "precio": 15000, "cantidad": 5},
    {"codigo": "P002", "nombre": "Mouse", "precio": 350, "cantidad": 20},
    {"codigo": "P003", "nombre": "Teclado", "precio": 800, "cantidad": 10},
    {"codigo": "P004", "nombre": "Monitor", "precio": 4500, "cantidad": 7}
]

print("=== PRODUCTOS DISPONIBLES ===")

for producto in inventario:
    print(producto)

codigo_buscar = "P002"

print("\n=== BÚSQUEDA DE PRODUCTO ===")

for producto in inventario:
    if producto["codigo"] == codigo_buscar:
        print("Producto encontrado:")
        print(producto)

valor_total = 0

for producto in inventario:
    subtotal = producto["precio"] * producto["cantidad"]
    valor_total += subtotal

print("\n=== VALOR TOTAL DEL INVENTARIO ===")
print("Lps.", valor_total)
```

```python
# Ejercicio 14: Registro de ventas
- Cree una lista de ventas.
- Cada venta debe tener cliente, producto, cantidad y precio unitario.
- Calcule el total por venta.
- Calcule el total general vendido.
- Identifique el cliente con la compra más alta.
```

```python
ventas = [
    {"cliente": "Juan", "producto": "Laptop", "cantidad": 1, "precio_unitario": 15000},
    {"cliente": "Maria", "producto": "Mouse", "cantidad": 3, "precio_unitario": 350},
    {"cliente": "Carlos", "producto": "Monitor", "cantidad": 2, "precio_unitario": 4500},
    {"cliente": "Ana", "producto": "Teclado", "cantidad": 4, "precio_unitario": 800}
]

print("=== TOTAL POR VENTA ===")

total_general = 0
cliente_mayor = ""
compra_mayor = 0

for venta in ventas:

    total_venta = venta["cantidad"] * venta["precio_unitario"]

    print(
        venta["cliente"],
        "-",
        venta["producto"],
        "- Total: Lps.",
        total_venta
    )

    total_general += total_venta

    if total_venta > compra_mayor:
        compra_mayor = total_venta
        cliente_mayor = venta["cliente"]

print("\n=== TOTAL GENERAL VENDIDO ===")
print("Lps.", total_general)

print("\n=== CLIENTE CON MAYOR COMPRA ===")
print(cliente_mayor, "- Lps.", compra_mayor)
```

```python
=== TOTAL POR VENTA ===
Juan - Laptop - Total: Lps. 15000
Maria - Mouse - Total: Lps. 1050
Carlos - Monitor - Total: Lps. 9000
Ana - Teclado - Total: Lps. 3200

=== TOTAL GENERAL VENDIDO ===
Lps. 28250

=== CLIENTE CON MAYOR COMPRA ===
Juan - Lps. 15000
```

```python
# Ejercicio 15: Limpieza básica de texto
- Defina o solicite una frase.
- Elimine espacios al inicio y al final.
- Convierta el texto a minúsculas.
- Reemplace espacios dobles por un solo espacio.
- Cuente cuántas palabras contiene.
- Explique los métodos strip(), lower(), replace() y split().
```

```python
frase = "   Hola   Mundo   de   Python   "

print("Frase original:")
print(frase)

frase_limpia = frase.strip()

frase_limpia = frase_limpia.lower()

while "  " in frase_limpia:
    frase_limpia = frase_limpia.replace("  ", " ")

palabras = frase_limpia.split()

cantidad_palabras = len(palabras)

print("\nFrase limpia:")
print(frase_limpia)

print("\nCantidad de palabras:")
print(cantidad_palabras)
```
